```python
import pandas as pd
data = pd.read_csv("motion_data_modified_2.csv")
```

```python
data["category"].unique()
```

```python
data.columns
```

```python
X = data.drop(['category','subject_id'], axis=1)
y = data["category"]
```

```python
y.unique()
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
class_to_encoded = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict([X_test.iloc[0]])
```

```python
class_to_encoded = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
```

```python
predicted_class_names = label_encoder.inverse_transform(y_pred)
```

```python
predicted_class_names
```
